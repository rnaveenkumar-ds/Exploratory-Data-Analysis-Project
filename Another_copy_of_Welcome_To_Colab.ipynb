# Welcome to Colab!

```
#Load DataSet
import pandas as pd
url = ("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df = pd.read_csv(url)
print("original dataset \n",df)
```

```
#inspect data
print(df.info())
print(df.describe())
```

```
#Handling Missing Values
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["PassengerId"] = df["PassengerId"].fillna(df["PassengerId"].mode()[0])
print("dataset after filling missing values \n",df["Age"])
print("dataset after filling missing values \n",df["PassengerId"])
```

```
#Remove Duplicates
df=df.drop_duplicates()
```

```
#Filter Data:Passangers in second class
df_second_class = df[df["PassengerId"] >= 4]
print("filtering the second class passangers list: \n",df_second_class)
```

```
import matplotlib.pyplot as plt
import seaborn as sns
#Bar chatt:survival rate
survival_rate =df.groupby("Pclass")["Survived"].mean()
survival_rate.plot(kind="bar",color="skyblue")
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```
#Histplot for Age
sns.histplot(df["Age"],color="orange",kde=True,edgecolor="green")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```
#Scatter plot: Age vs Fare
plt.scatter(df["Age"],df["Fare"],color="purple",marker="o",alpha=0.7)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
